# **Rotasi citra 90 derajat (aliasing)**

```
import numpy as np
from PIL import Image
import math
```

```
sudut_putar = 90
img = Image.open("pikachu.png")
img = img.resize((50,50))
img_array = np.array(img)  
```

```
img
```

```
img_array.shape
```

```
sudut_ubah=math.radians(sudut_putar)
kosinus=math.cos(sudut_putar)
sinus=math.sin(sudut_putar)
```

rumus rotasi.png

```
tinggi_baru = round(abs(50*kosinus)+abs(50*sinus))+1
lebar_baru = round(abs(50*kosinus)+abs(50*sinus))+1

titik_tengah_tinggi = round(((50+1)/2)-1)   
titiK_tengah_lebar = round(((50+1)/2)-1)

tengah_tinggi_baru = round(((tinggi_baru+1)/2)-1)
tengah_lebar_baru = round(((lebar_baru+1)/2)-1)
```

```
hasil_rotasi=np.zeros((tinggi_baru,lebar_baru,img_array.shape[2]))
```

```
for i in range(50):
    for j in range(50):
        y=50-1-i-titik_tengah_tinggi                  
        x=50-1-j-titiK_tengah_lebar                      

        y_baru=round(-x*sinus+y*kosinus)
        x_baru=round(x*kosinus+y*sinus)

        y_baru=tengah_tinggi_baru-y_baru
        x_baru=tengah_lebar_baru-x_baru

        if 0 <= x_baru < lebar_baru and 0 <= y_baru < tinggi_baru and x_baru>=0 and y_baru>=0:
            hasil_rotasi[y_baru,x_baru,:]=img_array[i,j,:]   
```

```
img_90derajat=Image.fromarray((hasil_rotasi).astype(np.uint8))
img_90derajat
```

```
import matplotlib.pyplot as plt

plt.figure(figsize=(10,10))
ax1=plt.subplot(2,2,1)
ax2=plt.subplot(2,2,2)
ax1.imshow(img)
ax1.set_title("Gambar awal")
ax2.imshow(img_90derajat)
ax2.set_title("Rotasi 90 derajat")
```
